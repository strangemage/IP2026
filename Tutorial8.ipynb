### Functions in Python

```python
def myfun(x, y):
    return x * y
myfun(3, 4)
```

```python
def myfun(x):
    return x * 3
def apply(q, x):
    return q(x)
apply(myfun, 7)
```

```python
apply(lambda z: z * 4, 7)
```

Default Values for Arguments

```python
def myfun(b, c=3, d="hello"):
    return b + c
myfun(5,3,"hello")

```

```python
myfun(5,3)
```

```python
myfun(5)
```

Keyword Arguments

```python
def myfun (a, b, c):
    return a-b
myfun(2, 1, 43)
```

```python
myfun(c=43, b=1, a=2)
```

```python
myfun(2, c=43, b=1)
```

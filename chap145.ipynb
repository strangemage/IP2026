### Morphological Transformations

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Erosion 이루전

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
erosion1 = cv2.erode(img,kernel,iterations = 1)
erosion2 = cv2.erode(img,kernel,iterations = 2)
erosion3 = cv2.erode(img,kernel,iterations = 3)

cv2.imshow('dst1',erosion1)
cv2.imshow('dst2',erosion2)
cv2.imshow('dst3',erosion3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
cv2.imshow('dst',dilation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
cv2.imshow('dst',opening)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
cv2.imshow('dst',closing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
cv2.imshow('dst',gradient)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
cv2.imshow('dst',tophat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
cv2.imshow('dst',blackhat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
